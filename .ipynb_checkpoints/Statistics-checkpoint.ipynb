```python
Statistics in Python

Statistics is used to analyze data, find patterns, and summarize information.

Python provides:

Built-in statistics module

NumPy

Pandas (for real datasets)

1ï¸âƒ£ Statistics using statistics module (Basic & Easy)
import statistics as st

data = [10, 20, 30, 40, 50]

âœ… Mean (Average)
print(st.mean(data))

âœ… Median (Middle value)
print(st.median(data))

âœ… Mode (Most frequent value)
print(st.mode(data))

âœ… Variance
print(st.variance(data))

âœ… Standard Deviation
print(st.stdev(data))

2ï¸âƒ£ Statistics using NumPy (Fast & Popular)
import numpy as np

data = np.array([10, 20, 30, 40, 50])

ğŸ“Œ Mean
np.mean(data)

ğŸ“Œ Median
np.median(data)

ğŸ“Œ Standard Deviation
np.std(data)

ğŸ“Œ Variance
np.var(data)

ğŸ“Œ Min & Max
np.min(data)
np.max(data)

3ï¸âƒ£ Statistics using Pandas (Real Data Analysis)
import pandas as pd

data = pd.Series([10, 20, 30, 40, 50])
```
