```python
#wapp  to print sum of first n even num , odd num
n = int(input("Enter n: "))

even = 0
odd= 0

for i in range(1,(2*n),1):
    if i%2==0:
        even=even+i
    else:
        odd=odd+i
print(even)
print(odd)
        
```

```python
#wapp to print count of even number and odd number blw 1000 and 2000
start=1000
end=2000
even=0
odd=0
while start <=end:
    if start%2 ==0:
      even=even+1
    else:
      odd=odd+1
    start=start+1
print(even)
print(odd)
```

```python
#wap to print reverse of any given number 
n=int(input("Enter any number "))
rev=0
while n!=0:
    dig=n%10
    rev=rev*10+dig
    n=n//10
print(rev)
```

```python
#wap to check wheather a given  number is an armstong no. or not
n=int(input("enter number"))
a=len(str(n))
sm=0
odd=n
while n!=0:
    dig = n  % 10
    sm = sm+(dig**a)
    n = n//10
if odd == sm:
    print("armstrong")
else:
    print("not armstrong") 
```
