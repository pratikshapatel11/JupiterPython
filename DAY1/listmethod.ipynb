```python
#find() return position of any character/substring
#strobj.find(character/substring,[start],[end])
s="welcome to bhopal"
print(s.find("o"))
print(s.find("o",5))
print(s.find("bhopal"))
print(s.find("z"))
```

```python
#index() return position of any character/string
#strobj.index(character/substring,[start],end])
s="welcome to bhopal"
print(s.index("o"))
print(s.find("o",5))
print(s.find("bhopal"))
print(s.find("z"))
```

```python
#count() -that return occurence of character in given string
#strobj.count(character,[start],[end])
s="welcome to city of lakes"
print(s.count("o"))
print(s.count("e"))
print(s.count("e",2,22))
print(s.count("e",2,5))
print(s.count("z"))
```
