```python
#wap to check wheather a given number is prime or not 
n = int(input("enter any number : "))
if n==0 or n==1:
    print(n,"not a prime")
elif n > 1:
    for i in range (2 , n ):
        if n % i ==0:
            print(n,"not a prime ")
            break
    else: 
        print(n , "prime number")
else :
    print(n,"not a prime")

    
```
